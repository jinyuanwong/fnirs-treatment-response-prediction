```python
import scipy.io
import numpy as np 

MAIN_FOLD = '/Users/shanxiafeng/Documents/Project/Research/fnirs-prognosis/code/fnirs-treatment-response-prediction/allData/diagnosis/matlab_process_MBLL'
WITH_age_path = MAIN_FOLD + '/with_age_data.mat'
NO_age_path = MAIN_FOLD + '/without_age_data.mat'
WITH_age = scipy.io.loadmat(WITH_age_path)['all_subject_data']
NO_age = scipy.io.loadmat(NO_age_path)['all_subject_data']
```

```python
# print(WITH_age)
print('WITH_age', WITH_age.shape)

# save them 
transpose_func = lambda data: np.transpose(data, (0, 2, 3, 1))

trans_WITH_age = np.nan_to_num(transpose_func(WITH_age), 0)
trans_NO_age = np.nan_to_num(transpose_func(NO_age), 0)

 # There are some strange points whose value is extremely large, so we need to remove them
trans_WITH_age[trans_WITH_age > 1] = 0
trans_NO_age[trans_NO_age > 1] = 0

trans_WITH_age[trans_WITH_age < -1] = 0
trans_NO_age[trans_NO_age < -1] = 0

np.save(MAIN_FOLD + '/with_age', trans_WITH_age)
np.save(MAIN_FOLD + '/without_age', trans_NO_age)
```
