    1: Scores of 2 (Female)
2. Age (continuous) 
3. Education (continuous)
4. Smoking (0/1)

5. Alcohol (0/1)

    1: Scores of 3 (current chronic alcohol)
6. Depression states (0/1)
7. HAMD scores (continuous)
8. Suicide risk (0/1)
    0: Scores of 0 or 1 (no risk or low risk).
    1: Scores of 2, 3, or 4 (moderate to high risk).

2. Ethinicity

```python
# Import packages and set the path of ipynb to the main folder

import numpy as np 
from scipy.io import loadmat
import sys 
import os
import pandas as pd 
def set_path():
    if sys.platform == 'darwin':
        print("Current system is macOS")
        main_fold_path = '/Users/shanxiafeng/Documents/Project/Research/fnirs-prognosis/code/fnirs-treatment-response-prediction'
    elif sys.platform == 'linux':
        print("Current system is Ubuntu")
        main_fold_path = '/home/jy/Documents/fnirs/treatment_response/fnirs-depression-deeplearning'
    else:
        print("Current system is neither macOS nor Ubuntu")
    os.chdir(main_fold_path)
    sys.path.append(main_fold_path)    

set_path()



subjects_table = pd.read_csv('Prerequisite/patients_info/dataset_514-Table 1.csv')
subjects_table.columns = subjects_table.columns.str.replace(' ', '')


# select target labels
target_label = ['ID', 'Gender', 'Age', 'Education', 'Smoking', 'Alcohol', 'TotalHAMD-17', 'HAMD3']
subjects_table = subjects_table[target_label] 

# add label of depression
subjects_table['Depression'] = subjects_table['ID'].apply(lambda x: 1 if 'DEP' in x or 'PT' in x else 0)

# return target labels
return_labels = ['Gender', 'Age', 'Education', 'Smoking', 'Alcohol', 'Depression', 'TotalHAMD-17', 'HAMD3']

for head_name in return_labels:
    arr_values = subjects_table[head_name].values
    print(head_name, arr_values) 
# subjects_table

# add 'Depression' label to the table is 'DEP' or 'PT' is in the ID, set the label to 1 else 0

education_dict = {
    1.: 0,
    2.: 6,
    3.: 10,
    4.: 12,
    5.: 13,
    6.: 16,
}

# Function to map education values, retaining original if not in dict
subjects_table['Education'] = subjects_table['Education'].map(lambda x: education_dict[x] if x in education_dict else x)
for head_name in return_labels:
    arr_values = subjects_table[head_name].values
    print(head_name, arr_values) 
```

```python
# Define which columns are categorical and which are continuous
categorical_columns = ['Gender', 'Education', 'Smoking', 'Alcohol', 'Depression']
continuous_columns = ['Age', 'TotalHAMD-17', 'HAMD3']

# Check for NaN values in each column
for col in subjects_table.columns:
    nan_count = subjects_table[col].isnull().sum()
    print(f"NaN count in {col}: {nan_count}")

# Replace NaN values in categorical columns with the mode
for col in categorical_columns:
    if col in subjects_table.columns:
        mode_value = subjects_table[col].mode()[0]
        print(f"Mode value for {col}: {mode_value}")
        subjects_table[col].fillna(mode_value, inplace=True)

# Calculate mean values for HCs and Depression subjects for continuous columns
hc_mean_values = subjects_table[subjects_table['Depression'] == 0][continuous_columns].mean()
depression_mean_values = subjects_table[subjects_table['Depression'] == 1][continuous_columns].mean()

# Replace NaN values in continuous columns with respective group means and convert to int
for col in continuous_columns:
    subjects_table.loc[(subjects_table['Depression'] == 0) & (subjects_table[col].isnull()), col] = int(hc_mean_values[col])
    subjects_table.loc[(subjects_table['Depression'] == 1) & (subjects_table[col].isnull()), col] = int(depression_mean_values[col])
    # Ensure the column is of integer type
    subjects_table[col] = subjects_table[col].astype(int)



# # Print the updated values for each column
# return_labels = ['Gender', 'Age', 'Education', 'Smoking', 'Alcohol', 'Depression', 'TotalHAMD-17', 'HAMD3']

print('-'*10)
print('After cleaning')
print('-'*10)
for col in subjects_table.columns:
    nan_count = subjects_table[col].isnull().sum()
    print(f"NaN count in {col}: {nan_count}")
    
for col in return_labels:
    subjects_table[col] = subjects_table[col].astype(int)    
    
for head_name in return_labels:
    arr_values = subjects_table[head_name].values
    print(head_name, arr_values)    
    
```

```python
median_age = subjects_table['Age'].median()
median_education = subjects_table['Education'].median()
# median_hamd = subjects_table['TotalHAMD-17'].median()
hamd_threshold = 8
```

```python
# finish return table     

# Transform columns according to the given criteria
subjects_table['Gender'] = subjects_table['Gender'].apply(lambda x: 0 if x == 1 else 1)
subjects_table['Smoking'] = subjects_table['Smoking'].apply(lambda x: 0 if x in [1,2] else 1)
subjects_table['Alcohol'] = subjects_table['Alcohol'].apply(lambda x: 0 if x in [1,2] else 1)
subjects_table['HAMD3'] = subjects_table['HAMD3'].apply(lambda x: 0 if x in [0] else 1)

# Including age education and HAMD-17 
subjects_table['Age'] = subjects_table['Age'].apply(lambda x: 1 if x >= median_age else 0)
subjects_table['Education'] = subjects_table['Education'].apply(lambda x: 1 if x >= median_education else 0)
subjects_table['TotalHAMD-17'] = subjects_table['TotalHAMD-17'].apply(lambda x: 1 if x >= hamd_threshold else 0)


# Rename columns to match the specified output names
subjects_table.rename(columns={'TotalHAMD-17': 'HAMD_Scores', 'HAMD3': 'Suicide_Risk'}, inplace=True)
```

```python
def one_hot_encode_labels(labels):
    encoded_labels = []
    for index, label in enumerate(labels.T):
        label = label.reshape(-1, 1)
        # print(label, label.shape)
        encoded = OneHotEncoder().fit_transform(label)
        encode = encoded.toarray()[..., np.newaxis]
        encoded_labels.append(encode)
        
    concate_labels = np.concatenate(encoded_labels, axis=-1)
    transpose_labels = np.transpose(concate_labels, (0, 2, 1))
    return transpose_labels

res = one_hot_encode_labels(multi_task_label_one_hot_encoded)
np.save('allData/diagnosis514/multi_task_label_5_one_hot_encoded.npy', res)
```
